```python
from scenedetect import AdaptiveDetector, detect
from tqdm.auto import tqdm
import pandas as pd
import os

for video in os.listdir(r'E:\Downloads\atm24949_withsound\video'):
    if video.endswith('.mp4'):
        print(video)
        scene_list = detect(fr'E:\Downloads\atm24949_withsound\video\{video}', AdaptiveDetector(adaptive_threshold=10, min_scene_len=25))

        df = pd.DataFrame(columns=['start_frame', 'end_frame', 'start_time', 'end_time', 'fps'])

        for i, (x, y) in enumerate(scene_list):
            df.loc[i] = [x.frame_num, y.frame_num, x.get_seconds(), y.get_seconds(), x.framerate]
        df.to_csv('./scenedetect/'+video[:-4]+'_scenedetect.csv', index=False)
```
