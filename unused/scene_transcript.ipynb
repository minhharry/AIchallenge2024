```python
import pandas as pd

df = pd.read_csv('scenedetect.csv')

df['text'] = pd.Series(['None' for i in range(len(df))])
df['text_EN'] = pd.Series(['None' for i in range(len(df))])
```

```python
import json
OFFSET = 1
with open('L01_V001.json', 'r', encoding='utf-8') as f:
    transcript = json.load(f)
    for index, row in df.iterrows():
        temp = ''
        for segment in transcript:
            if segment['start'] >= row['start_time'] - OFFSET and segment['start'] + segment['duration'] <= row['end_time'] + OFFSET:
                temp += segment['text'] + ' '
        df.at[index, 'text'] = temp

with open('L01_V001_EN.json', 'r', encoding='utf-8') as f:
    transcript = json.load(f)
    for index, row in df.iterrows():
        temp = ''
        for segment in transcript:
            if segment['start'] >= row['start_time'] - OFFSET and segment['start'] + segment['duration'] <= row['end_time'] + OFFSET:
                temp += segment['text'] + ' '
        df.at[index, 'text_EN'] = temp
```
