```python
import pandas as pd

```

```python
import json
import glob
```

```python
file_paths = glob.glob('media-info/*')
file_paths
```

```python
df = pd.DataFrame()
for file in file_paths:
    with open(file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    data['video_id'] = data['watch_url'].split('=')[-1]
    data['filename'] = file.split('\\')[-1]
    df = pd.concat([df, pd.DataFrame([data])], ignore_index=True)
```

```python
df.to_csv('media_info.csv', index=False)
```
